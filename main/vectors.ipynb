```python
#import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
vec = np.array([[0, 0, 2, 6], [0, 0, 5, 2]])

X, Y, U, V = zip(*vec)
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color = 'red', alpha = .6)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_title("Plotting Vectors in R2")
ax.set_xlabel('x-axis', fontsize =16)
ax.set_ylabel('y-axis', fontsize =16)
ax.grid()
ax.text(5, 2, '$(5, 2)$', fontsize = 16)
ax.text(2, 6, '$(2, 6)$', fontsize = 16)


ax.text(2, 4, '$\sqrt{2^2+6^2}$', fontsize = 16, rotation = np.arctan(7/4)*180/np.pi)
ax.text(3, 1.9, '$\sqrt{5^2+2^2}$', fontsize = 16, rotation = np.arctan(6/8)*180/np.pi ) 
# use arctan to get the rotation degree and convert from radian to degrees
plt.show()
```

```python
vec = np.array([[0, 0, 0, 1, 1, 1], 
                [0, 0, 0, -2, 2, 5], 
                [0, 0, 0, 3, -2, 1]])
X, Y, Z, U, V, W = zip(*vec)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, Z, U, V, W, length=1, normalize=False, color = 'red', alpha = .6,arrow_length_ratio = .18, pivot = 'tail',
          linestyles = 'solid',linewidths = 3)
ax.grid()
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([0, 5])

ax.text(1, 1, 1, '$(1, 1, 1)$')
ax.text(-2, 2, 5, '$(-2, 2, 5)$')
ax.text(3, -2, 1, '$(3, -2, 1)$')
ax.text(0, 0, 0, '$(0, 0, 0)$')

point1 = [1, 1, 1]
point2 = [1, 1, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [0, 0, 0]
point2 = [1, 1, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [3, -2, 1]
point2 = [3, -2, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [0, 0, 0]
point2 = [3, -2, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [-2, 2, 5]
point2 = [-2, 2, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [0, 0, 0]
point2 = [-2, 2, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

plt.show()
```
